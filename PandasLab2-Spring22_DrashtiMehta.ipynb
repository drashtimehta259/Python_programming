QUESTION 1C: What is the mean earning of students 10years after enrollment of those schools?

```python
import numpy as np
sort_50['MD_EARN_WNE_P10'] = sort_50['MD_EARN_WNE_P10'].replace(['PrivacySuppressed'],np.nan)
sort_50['MD_EARN_WNE_P10'].dropna()
```
