```python
# to check and note the on going process, we should save it as a log file

fasta_log = open('pd_hw_log.log', 'w')
fasta_log.truncate(0)

# empty dictionary is created
empty_dictionary = dict()
count_seq = 0


#start and stop codons
start_codon_list = ['ATG']
stop_codon_list = ['TGA', 'TAA', 'TAG']


# open and read given fasta file using the parser function and empty dictionary
try:
    for empty_dictionary in readingFASTAFile("D:/UNCC/Sem_2/6112_python/Mdomestica-pandasLab.fa"):
        count_seq = count_seq + 1
except:
    print("Cannot read the file")

print("sequence length = " +str(count_seq))



# to check the given conditions mentioned in the question, we should create list variables
indexes = list()
seqLen = list()
GC_perc = list()
valid_start = list()
valid_stop = list()
completeFrame = list()


# appending values

for key,value in empty_dictionary.items():
    
    #length variable to store length value:
    sequence_length = len(value)
    print("sequence length = " +str(sequence_length))
    
    #GC content:
    GC_content = (value.count('G') + value.count('C')) / sequence_length
    #print("GC content = " +str(GC_content))
    GC_percent = ((value.count('G') + value.count('C')) / sequence_length) * 100
    print("GC percentage = " +str(GC_percent))
    
    #start codon
    start_codon = str(value)[0:3]
    print("start codon = " +str(start_codon))
    
    #stop codon
    stop_codon = str(value)[sequence_length - 3:]
    print("stop codon = " +str(stop_codon))
    
    #complete frame
    print("Frame validation = " +str(validCodonFrame(value)))
    
    # writing log files
    fasta_log.write("Sequence length = " +str(sequence_length))
    fasta_log.write("\nGC percentage = " +str(GC_percent))
    fasta_log.write("\nStart codon = " +str(start_codon))
    fasta_log.write("\nStop codon = " +str(stop_codon))
    fasta_log.write("\nFrame validation = " +str(validCodonFrame(value)))
    print()
    
    # appending to the list 
    indexes.append(key[0:4])
    seqLen.append(sequence_length)
    GC_perc.append(GC_percent)
    if start_codon in start_codon_list:   
        valid_start.append(True)
    else:
        valid_start.append(False)
    if stop_codon in stop_codon_list:
        valid_stop.append(True)
    else:
        valid_stop.append(False)
    completeFrame.append(validCodonFrame(value))

# creating result dictionary
Result = {
    
    "Seq length" : seqLen,
    "GC percent" : GC_perc,
    "Start_validation" : valid_start,
    "Stop_validation" : valid_stop,
    "Complete frame" : completeFrame
}
    
```

```python
# as required, we plot a table here. Inputting it into log file and priting

sequence_result_table = pd.DataFrame(Result, index = indexes)
fasta_log.write(str(sequence_result_table))
fasta_log.close()
sequence_result_table
```
