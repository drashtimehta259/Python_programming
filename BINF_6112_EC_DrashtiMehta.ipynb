# The main purpose of this assignment is to check the status if business on getting affected by covid. Which industry provides maximum employment and which industry has fallen down due to the onset of the pandemic will be discussed in this assignment.

```python
# importing pandas and matplotlib
import pandas as pd
import matplotlib as plt

# setting path for the files
path = "D:/UNCC/Sem_2/6112_python/CSV/"

# assigning the files
confirmed_cases = pd.read_csv(path + "time_series_covid19_confirmed_US.csv")
business = pd.read_csv(path + "County_Business_Patterns_-_States_2019.csv")
```

```python
# grouping by states
confirmed = confirmed_cases.groupby("Province_State", as_index = False).sum()
confirmed
```

```python
# checking monthly increase

confirmed['february_increase'] = confirmed["2/29/2020"] - confirmed["2/1/2020"]
confirmed['march_increase'] = confirmed["3/31/2020"] - confirmed["3/1/2020"]
confirmed['april_increase'] = confirmed["4/30/2020"] - confirmed["4/1/2020"]
confirmed['may_increase'] = confirmed["5/31/2020"] - confirmed["5/1/2020"]
confirmed['june_increase'] = confirmed["6/30/2020"] - confirmed["6/1/2020"]
confirmed['july_increase'] = confirmed["7/31/2020"] - confirmed["7/1/2020"]
confirmed['august_increase'] = confirmed["8/31/2020"] - confirmed["8/1/2020"]
confirmed['september_increase'] = confirmed["9/30/2020"] - confirmed["9/1/2020"]
confirmed['october_increase'] = confirmed["10/31/2020"] - confirmed["10/1/2020"]
confirmed
```

```python
# checking what is the increase in the amount

feb_max = confirmed["february_increase"].max()
print("February increase: " +str(feb_max))
march_max = confirmed["march_increase"].max()
print("March increase: " +str(march_max))
april_max = confirmed["april_increase"].max()
print("April increase: " +str(april_max))
may_max = confirmed["may_increase"].max()
print("May increase: " +str(may_max))
june_max = confirmed["june_increase"].max()
print("June increase: " +str(june_max))
july_max = confirmed["july_increase"].max()
print("July increase: " +str(july_max))
august_max = confirmed["august_increase"].max()
print("August increase: " +str(august_max))
september_max = confirmed["september_increase"].max()
print("September increase: " +str(september_max))
october_max = confirmed["october_increase"].max()
print("October increase: " +str(october_max))
```

# We can see that the maximum increase seems to be in the month of July. We all know that the cases started appearing in the month of February and peaking in March. The lockdown in the US was almost from mid March to early July which is taken into consideration here. In July, everything opened again slowly and gradually and that might be the reason where we can see the cases peaking up again. Hence, some more months are taken into consideration to see decline in number of cases.

```python
# mean values of monthly increase
confirmed['average_increase'] = confirmed[['february_increase', 'march_increase', 'april_increase', 'may_increase', 'june_increase', 'july_increase', 'august_increase', 'september_increase', 'october_increase']].mean(axis=1)
confirmed
```

```python
# checking which state have the maximum increase in each month

sorting_increase_feb = confirmed.sort_values(by = ['february_increase'], ascending = False)
highest_increase_feb = sorting_increase_feb["Province_State"].iloc[0]
print("Highest February Increase: " +str(highest_increase_feb))

sorting_increase_march = confirmed.sort_values(by = ['march_increase'], ascending = False)
highest_increase_march = sorting_increase_march["Province_State"].iloc[0]
print("Highest March Increase: " +str(highest_increase_march))

sorting_increase_april = confirmed.sort_values(by = ['april_increase'], ascending = False)
highest_increase_april = sorting_increase_april["Province_State"].iloc[0]
print("Highest April Increase: " +str(highest_increase_april))

sorting_increase_may = confirmed.sort_values(by = ['may_increase'], ascending = False)
highest_increase_may = sorting_increase_may["Province_State"].iloc[0]
print("Highest May Increase: " +str(highest_increase_may))

sorting_increase_june = confirmed.sort_values(by = ['june_increase'], ascending = False)
highest_increase_june = sorting_increase_june["Province_State"].iloc[0]
print("Highest June Increase: " +str(highest_increase_june))

sorting_increase_july = confirmed.sort_values(by = ['july_increase'], ascending = False)
highest_increase_july = sorting_increase_july["Province_State"].iloc[0]
print("Highest July Increase: " +str(highest_increase_july))

sorting_increase_august = confirmed.sort_values(by = ['august_increase'], ascending = False)
highest_increase_august = sorting_increase_august["Province_State"].iloc[0]
print("Highest August Increase: " +str(highest_increase_august))

sorting_increase_september = confirmed.sort_values(by = ['september_increase'], ascending = False)
highest_increase_september = sorting_increase_september["Province_State"].iloc[0]
print("Highest September Increase: " +str(highest_increase_september))

sorting_increase_october = confirmed.sort_values(by = ['october_increase'], ascending = False)
highest_increase_october = sorting_increase_october["Province_State"].iloc[0]
print("Highest October Increase: " +str(highest_increase_october))
```

```python
# highest monthly average state

sorting_increase_average = confirmed.sort_values(by = ['average_increase'], ascending = False)
highest_increase_average = sorting_increase_average["Province_State"].iloc[0]
print("Highest Average Increase: " +str(highest_increase_average))
```
